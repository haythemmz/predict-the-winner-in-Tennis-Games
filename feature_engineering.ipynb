Dans le code suivant , nous avons développé une fonction qui prend comme paramètre une dataframe et retourne une dataframe qui résume nombre et le pourcentage des données manquant pour chaque variable 

```python
def missing_data_function(frame:

    total = frame.isnull().sum().sort_values(ascending=False)
    percent = (frame.isnull().sum()*100 / frame.isnull().count()).sort_values(ascending=False)
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    return missing_data
missing_data=missing_data_function(stat)
#missing_data
```

```python
#  la différence d’ACE par jeux entre les deux joueurs 
def ace(id_1, id_2, stat):
    o1=0
    s1=0
    o2 = 0
    s2 = 0
    for i, j , k ,l in zip(stat['ID1'],stat['ID2'],stat['ACES_1'],stat['ACES_2']):
        if i == id_1 :
            s1+=k
            o1+=1
        if j == id_1:
            s1+=l
            o1+=1
        if i == id_2:
            s2 += k
            o2 += 1
        if j == id_2:
            s2 += l
            o2+= 1
    ace1=s1/o1
    ace2=s2/o2
    return ace1-ace2
```

```python
# la différence des double faute  par jeux   entre les deux joueurs
def averge_DF(id_1,id_2,stat):
    o1=0
    s1=0
    o2=0
    s2=0
    for i,j,k,l in zip(stat['ID1'], stat['ID2'], stat['DF_1'], stat['DF_2']):
        if i == id_1 :
            s1+=k
            o1+=1
        if j == id_1:
            s1+=l
            o1+=1
        if i == id_2:
            s2 += k
            o2 += 1
        if j == id_2:
            s2 += l
            o2 += 1
    
    
    df1=s1/o1
    df2=s2/o2
    
    return df1-df2
```

```python
# différence entre la somme  Nombre de points gagnés sur premier services sur Nombre de points joués sur premier service 
# sur nombre des matchs jouées
def PFS (id_1,id_2,stat):
    o1=0
    s1=0
    s2=0
    o2=0
    for i,j,k,l,m,n in zip(stat['ID1'],stat['ID2'],stat['W1S_1'],stat['W1SOF_1'],stat['W1S_2'],stat['W1S_2']):
        if i == id_1 :
            if l !=0:
                s1+= k/l
            o1+=1
        if j== id_1:
            if n !=0:
                s1+=m/n
            o1+=1

        if i == id_2:
            if l !=0:
                s2 += k/l
            o2 += 1
        if j == id_2:
            if n !=0:
                s2 += m/n
            o2 += 1
    fs1=s1/o1
    fs2=s2/o2
    return fs1-fs2
```

```python
l=[]
for i,j in zip(df['ID1_G'],df['ID2_G']):
    l.append(PFS(i,j,stat))
df['DF_PFS']=l
```

```python
# différence entre la somme  Nombre de points gagnés sur seconde services sur Nombre de points joués sur seconde service 
# sur nombre des matchs jouées
def PSS (id_1,id_2,stat):
    o1=0
    s1=0
    s2=0
    o2=0
    for i,j,k,l,m,n in zip(stat['ID1'],stat['ID2'],stat['W2S_1'],stat['W2SOF_1'],stat['W2S_2'],stat['W2S_2']):
        if i == id_1 :
            if l !=0:
                s1+= k/l
            o1+=1
        if j== id_1:
            if n !=0: 
                s1+=m/n
            o1+=1

        if i == id_2:
            if l !=0:
                s2 += k/l
            o2 += 1
        if j == id_2:
            if n !=0:
                s2 += m/n
            o2 += 1
    fs1=s1/o1
    fs2=s2/o2
    return fs1-fs2
```

```python
l=[]
for i,j in zip(df['ID1_G'],df['ID2_G']):
    l.append(PSS(i,j,stat))
df['DF_PSS']=l
```

```python
#différence entre nombre des matchs gagné sur le nombre des matchs jouées
def win (id_1,id_2,stat):
    w1=0
    o1=0
    w2=0
    o2=0
    for i,j in zip(stat['ID1'],stat['ID2']):
        if i==id_1:
            w1+=1
            o1+=1
        if j== id_1:
            o1+=1
        if i==id_2:
            w2+=1
            o2+=1
        if  j==id_2:
            o2+=1
    win1=w1/o1
    win2=w2/o2
    return win1-win2
```

```python
l=[]
for i,j in zip(df['ID1_G'],df['ID2_G']):
    l.append(win(i,j,stat))
df['WIN']=l
```

```python
#différence entre les breaks gagné parmi les breaks joués  
def breack (id_1,id_2,stat): 
    o1=0
    s1=0
    s2=0
    o2=0
    for i,j,k,l,m,n in zip(stat['ID1'],stat['ID2'],stat['BP_1'],stat['BPOF_1'],stat['BP_2'],stat['BPOF_2']):
        if i == id_1 :
            if l!=0:
                s1+= k/l
            o1+=1
        if j== id_1:
            if n!=0:
                s1+=m/n
            o1+=1

        if i == id_2:
            if l!=0:
                s2 += k/l
            o2 += 1
        if j == id_2:
            if n!=0:
                s2 += m / n
            o2 += 1
    fs1=s1/o1
    fs2=s2/o2
    return fs1-fs2
```

```python
l=[]
for i,j in zip(df['ID1_G'],df['ID2_G']):
    l.append(breack(i,j,stat))
df['Break']=l
```

```python
# l'historique des matchs entre les deux joueus 
def head_to_head(id_1,id_2,stat):
    totale=0
    P1_win=0
    P2_win=0
    for i,j in zip(stat['ID1'], stat['ID2']):
        if i==id_1 and j==id_2: # le joueur J1 gagne 
            P1_win+=1
            totale+=1
        if j==id_1 and i==id_2: # le joueur J2 gagne 
            P2_win+=1
            totale+=1

        try:
            return (P1_win-P2_win)/totale
        except:
            return 0
```

```python
l=[]
for i,j in zip(df['ID1_G'],df['ID2_G']):
    l.append(head_to_head(i,j,stat))
df['H2H']=l
```

La fonction suivante décrit les compertements de deux joueurs vis à vis leurs adversaires communs 

```python
def win_common_opponent(id_1,id_2,stat):
    oppent_1=[] # liste des adversaires du premier joueur
    oppent_2=[] # liste des adversaires du second joueur
    for i,j in zip(stat['ID1'],stat['ID2']):
        if i == id_1 :
            oppent_1.append(j)
        if j == id_1:
            oppent_1.append(i)
        if i == id_2 :
            oppent_2.append(j)
        if j == id_2:
            oppent_2.append(i)

    intersection=np.intersect1d(oppent_1,oppent_2)
    dio_1= []
    dio_2=[]
    for i in intersection :
        for k,l in zip(stat['ID1'],stat['ID2']):
            if k == id_1 and l == i :
                dio_1.append(1)
            if k==i and l==id_1:
                dio_1.append(0)
            if k == id_2 and l == i:
                oppent_2.append(1)
            if k == i and l == id_2:
                oppent_2.append(0)




    wrp_1=np.mean(dio_1)
    wrp_2=np.mean(dio_2)
    return wrp_1-wrp_2
```

```python
l=[]
for i,j in zip(df['ID1_G'],df['ID2_G']):
    l.append(win_common_opponent(i,j,stat))
df['CO']=l
```

```python
# la différence de points entre les deux joueurs 
def point_def(id_1,id_2,player_rates):
    
    if id_1 in list(player_rates['ID_P_R']):
        rank_1=player_rates[player_rates['ID_P_R'] == id_1]['POINT_R'].values[0]

    else: 
        rank_1= min(list(player_rates['POINT_R'])) # si le joueurs est non classé on l'attribue le minimum de point  

    if  id_2 in list(player_rates['ID_P_R']):
        rank_2 = player_rates[player_rates['ID_P_R'] == id_2]['POINT_R'].values[0]
    else:
        rank_2= min(list(player_rates['POINT_R'])) # si le joueurs est non classé on l'attribue le minimum de point  
        
    return rank_1-rank_2
```

```python
l=[]
for i, j in zip(df['ID1_G'],df['ID2_G']):
    l.append(point_def(i,j,player_rates))
df['DF_POINT']=l
```

```python
# la différence de rangs  entre les deux joueurs 
def rank_def(id_1,id_2,player_rates):
    
    if id_1 in list(player_rates['ID_P_R']):
        rank_1=player_rates[player_rates['ID_P_R'] == id_1]['POS_R'].values[0]

    else: 
        rank_1= max(list(player_rates['POS_R']))  # si le joueurs est non classé on l'attribue le rang de dernier 

    if  id_2 in list(player_rates['ID_P_R']):
        rank_2 = player_rates[player_rates['ID_P_R'] == id_2]['POS_R'].values[0]
    else:
        rank_2= max(list(player_rates['POS_R'])) # si le joueurs est non classé on l'attribue le rang de dernier 
        
    return rank_1-rank_2
```

```python
l=[]
for i, j in zip(df['ID1_G'],df['ID2_G']):
    l.append(rank_def(i,j,player_rates))
df['DF_RANK']=l
```

```python
# modèlisation de la " complétude " des joueurs on dit qu'un joueur est complet si il est à la fois un bon defensur
#  et  un bon attaquant , on modèlise l'attaque par " Ace par matchs " et la défense par
# "breaks gagnés par break joués par matchs" la variable est la résultante de la production 
# 
def complet (id_1,id_2,stat): 
    o1=0
    s1=0
    a1=0
    s2=0
    o2=0
    a2=0
    for i, j ,k ,l,m,n,y,z in zip(stat['ID1'],stat['ID2'],stat['BP_1'],stat['BPOF_1'],stat['BP_2'],stat['BPOF_2'],stat['ACES_1'],stat['ACES_2']):
        if i == id_1 :
            if l!=0:
                s1+= k/l
            a1+=y
            o1+=1
        if j== id_1:
            if n!=0:
                s1+=m/n
            o1+=1
            a1+=z

        if i == id_2:
            if l!=0:
                s2 += k/l
            a2+=y
            o2 += 1
        if j == id_2:
            if n!=0:
                s2 += m / n
            a2+=z
            o2 += 1
    
    fs1=s1*a1/o1
    fs2=s2*a2/o2
    return fs1-fs2
```

```python
l=[]
for i, j in zip(df['ID1_G'],df['ID2_G']):
    l.append(complet(i,j,stat))
df['complet']=l
```
